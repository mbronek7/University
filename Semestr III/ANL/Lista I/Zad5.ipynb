```ruby
require 'nyaplot'
class Function
 
  def initialize(&block)
     @function = block
  end

  def value (x)   
    @function.call x
  end
 
  
 def plot a, b
   x = []; y = []; 
   x = (a..b) .collect { |v| v.to_f }
   y = x.collect { |v| value(v)}
   plot1 = Nyaplot::Plot.new
   plot1.add(:line, x, y)
   plot1.show
end

def zero_of_fun(a,b,c)
   begin 
   delta = b*b-4*a*c 
   x1 = (-b + Math.sqrt(delta)) / (2.0*a)
   x2 = (-b - Math.sqrt(delta)) / (2.0*a)
   print "Miejsca zerowe to: x1 = #{x1} || x2 = #{x2}\n"
   print"WartoÅ›Ä‡ funkcji w tych punktach to odpowiednio dla #{x1} -- #{@function.call x1}, a dla #{x2} --#{@function.call x2}\n "
   rescue 
   print "delta < 0"
   end
end
  
end

#
#
#
#
def calc(x,k)
    sum = 0 
    for i in 1..k 
        sum+=(((-1) ** (i-1) ) * (((x-1) ** i) / i))
    end
    sum
end

def ln x
    if  x > 5
      temp = x / Math::E 
     return 1 + 1 + ln(temp/Math::E)
    else
     return 1 + calc((x / Math::E),30)
    end 
end
```

```ruby
mojaimplementacja = Function.new{|x| ln x}
mojaimplementacja.plot 1,100
```
