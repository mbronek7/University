```ruby
require 'nyaplot'
require 'gnuplot'
class Function
 
  def initialize(&block)
     @function = block
  end

  def value (x)   
    @function.call x
  end
 
  
 def plot a, b
   x = []; y = []; 
   x = (a..b) .collect { |v| v.to_f }
   y = x.collect { |v| value(v)}
   plot1 = Nyaplot::Plot.new
   plot1.add(:line, x, y)
   plot1.show
end

 
def plot2 a, b

    Gnuplot.open do |gp|
     Gnuplot::Plot.new( gp ) do |plot|

         
         plot.xrange "[#{a}:#{b}]"
         plot.ylabel "y"
         plot.xlabel "x"

         x = (a..b) .collect { |v|v.to_f }
         y = x.collect { |v| value(v)}

         plot.data << Gnuplot::DataSet.new( [x, y] ) do |ds|
          ds.with = "lines"
         end


     end
    end
end  
  
def zero_of_fun(a,b,c)
   begin 
   delta = b*b-4*a*c 
   x1 = (-b + Math.sqrt(delta)) / (2.0*a)
   x2 = (-b - Math.sqrt(delta)) / (2.0*a)
   print "Miejsca zerowe to: x1 = #{x1} || x2 = #{x2}\n"
   print"Wartość funkcji w tych punktach to odpowiednio dla #{x1} -- #{self.value(x1)}, a dla #{x2} --#{self.value(x1)}\n "
   rescue 
   print "delta < 0"
   end
end
  
end
```

```ruby
function = Function.new{|x| 9.0*x**2 + (20 ** 8)*x + 1.0}
function.plot -100,200
function.zero_of_fun(9.0,  20 ** 8,   1.0)
```

```ruby
function = Function.new{|x| 1.0*x**2 + 14782*x + 1.123e-9}
function.plot -100,200
function.zero_of_fun(1.0,  14782,     1.123e-9)
```
